###Here are the explaination of each column

Crop: This column likely contains the name or type of the crop that is being cultivated or harvested. For example, it could include names like wheat, rice, corn, soybeans, etc.

Crop_Year: This column likely represents the year in which the crop was planted or harvested. It helps to track the timeline of agricultural activities for each specific crop.

Season: This column likely represents the season in which the crop was grown or harvested. Seasons can include categories like "summer," "winter," "monsoon," etc., depending on the region.

State: This column likely represents the state or geographical region where the crop was cultivated. It helps to provide information about the location of the agricultural activity.

Area: This column likely represents the area of land in a specific unit (such as hectares or acres) where the crop was planted or harvested.

Production: This column likely represents the quantity or amount of the harvested crop. It indicates how much of the crop was successfully grown and harvested.

Annual_Rainfall: This column likely represents the amount of rainfall that occurred during the year in the specific region. Rainfall is an important factor in agricultural productivity as it affects crop growth.

Fertilizer: This column likely represents information about the type or amount of fertilizers used during the cultivation of the crop. Fertilizers are used to enrich the soil with essential nutrients for plant growth.

Pesticide: This column likely represents information about the type or amount of pesticides used to protect the crop from pests and diseases.

Yield: This column likely represents the yield of the crop, which is the production amount divided by the cultivated area. It provides insight into the efficiency and productivity of the agricultural process.

# CA 1 Question Answers

1. What is the most common crop in the dataset?

```python
most_common_crop = df['Crop'].mode()[0]
print("The most common crop in the dataset is:", most_common_crop)
```

2. Which year had the highest crop production?

```python
year_with_highest_production = df[df['Production'] == df['Production'].max()]['Crop_Year'].values[0]

print("The year with the highest crop production is:", year_with_highest_production)
```

3. What is the distribution of different seasons in the dataset?

```python
season_distribution = df['Season'].value_counts()


print("Distribution of Seasons:")
print(season_distribution)
```

4. Does the choice of crop season impact crop production?

```python
season_production = df.groupby('Season')['Production'].mean()

print("Average Production by Season:")
print(season_production)
```

5. What is the range of annual rainfall in the dataset?

```python
min_rainfall = df['Annual_Rainfall'].min()
max_rainfall = df['Annual_Rainfall'].max()

rainfall_range = max_rainfall - min_rainfall

print("Range of Annual Rainfall:", rainfall_range)
```

6. How many unique fertilizer types are present in the dataset?

```python
unique_fertilizer_count = df['Fertilizer'].nunique()


print("Number of Unique Fertilizer Types:", unique_fertilizer_count)
```

7. What is the most commonly used pesticide

```python
pesticide_counts = df['Pesticide'].value_counts()


most_common_pesticide = pesticide_counts.index[0]


print("Most Commonly Used Pesticide:", most_common_pesticide)
```

8. What is the distribution of crop yields?

```python
yield_stats = df['Yield'].describe()


plt.figure(figsize=(10, 6))
plt.hist(df['Yield'], bins=20, edgecolor='k')
plt.title('Distribution of Crop Yields')
plt.xlabel('Yield')
plt.ylabel('Frequency')
plt.grid(True)

print("Summary Statistics for Crop Yields:")
print(yield_stats)

plt.show()





```

9. Is there a correlation between annual rainfall and crop production?

```python
correlation = df['Annual_Rainfall'].corr(df['Production'])


if correlation > 0:
    correlation_result = "There is a positive correlation."
elif correlation < 0:
    correlation_result = "There is a negative correlation."
else:
    correlation_result = "There is no correlation."


print("Correlation Coefficient:", correlation)
print("Correlation Result:", correlation_result)
```

10. Does the choice of crop season impact crop production?

```python
season_production = df.groupby('Season')['Production'].mean().sort_values()


print("Average Production by Season:")
print(season_production)
```

11. Is there a relationship between crop area and pesticide use?

```python
correlation = df['Area'].corr(df['Pesticide'])

if correlation > 0:
    relationship_result = "There is a positive relationship."
elif correlation < 0:
    relationship_result = "There is a negative relationship."
else:
    relationship_result = "There is no relationship."


print("Correlation Coefficient between Area and Pesticide:", correlation)
print("Relationship Result:", relationship_result)
```

12. How does crop yield vary with different types of crops?

```python
crop_yield = df.groupby('Crop')['Yield'].mean().sort_values()


plt.figure(figsize=(12, 6))
crop_yield.plot(kind='bar', color='green')
plt.title('Average Crop Yield by Crop Type')
plt.xlabel('Crop Type')
plt.ylabel('Average Yield')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.show()
```

13. Are there any noticeable trends in crop production over the years?

```python
yearly_production = df.groupby('Crop_Year')['Production'].sum()


plt.figure(figsize=(12, 6))
yearly_production.plot(marker='o', linestyle='-', color='blue')
plt.title('Crop Production Over the Years')
plt.xlabel('Crop Year')
plt.ylabel('Total Production')
plt.grid(True)


plt.show()
```

14. What is the relationship between fertilizer and pesticide usage?

```python
correlation = df['Fertilizer'].corr(df['Pesticide'])


if correlation > 0:
    relationship_result = "There is a positive relationship."
elif correlation < 0:
    relationship_result = "There is a negative relationship."
else:
    relationship_result = "There is no relationship."


print("Correlation Coefficient between Fertilizer and Pesticide:", correlation)
print("Relationship Result:", relationship_result)
```

15. What is the distribution of crop production across different states

```python
state_production = df.groupby('State')['Production'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
state_production.plot(kind='bar', color='purple')
plt.title('Crop Production Across States')
plt.xlabel('State')
plt.ylabel('Total Production')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

16.Do certain states or regions consistently outperform others in terms of crop production? 

```python
production_stats = df.groupby('State')['Production'].describe()
print(production_stats)
```

```python

plt.figure(figsize=(12, 6))
production_stats['mean'].plot(kind='bar', color='orange')
plt.title('Average Crop Production by State')
plt.xlabel('State')
plt.ylabel('Average Production')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

17.What is the total production of all crops combined for each year?

```python
yearly_total_production = df.groupby('Crop_Year')['Production'].sum()

print("Total Production for Each Year:")
print(yearly_total_production)
```

18.Do different regions or states have distinct preferences for specific crop types?

```python
region_crop_counts = df.groupby(['State', 'Crop'])['Crop'].count().unstack().fillna(0)


print(region_crop_counts)
```

```python
plt.figure(figsize=(30, 16))
region_crop_counts.plot(kind='bar', stacked=True)
plt.title('Crop Type Distribution by State/Region')
plt.xlabel('State/Region')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Crop Type', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
